```python
import pandas as pd
```

```python
df=pd.read_csv('moru_reviews_static.csv')
print(df.head())
```

```python
import re
import emoji

def clean_text(text):
    text = str(text).lower()
    text = emoji.replace_emoji(text, replace='')
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'[^a-zA-Z\u0900-\u097F\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['clean_review'] = df['review_text'].apply(clean_text)
print(df['clean_review'].head())
print("Total reviews after cleaning:", len(df))
df.columns

```

```python
def rating_sentiment(rating):
    if rating >= 4:
        return "Good"
    elif rating == 3:
        return "Neutral"
    else:
        return "Bad"

df['baseline_sentiment'] = df['rating'].apply(rating_sentiment)
```

```python
df.to_csv('moru_reviews_cleaned_static.csv', index=False)
```
